Задание 2. Работа с выбросами

```python
# СКО в столбце "Age". Проверим столбtw на выбросы.
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['Age'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
'''
В столбце "Age" найдены выбросы.
В столбце могут быть только следующие значения: 1 или 2. Так как 2 подразумевается в таблице, но её там нет, возможно, 
выбросы - это ошибки ввода данных, и вместо 9 следовало бы расставить 2. 
Выбросов не много и на расчет моды они не влияют, данные оставаляю в исходном виде.
'''
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

```python
q1 = df['Rectal temperature'].quantile(0.25)
q3 = df['Rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['Rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
'''
В столбце "Rectal temperature" найдены выбросы.
Максимальная и минимальная температура не выходят за логические пределы - ошибок нет.
На качестве анализа выборсы не сказываются. Данные оставляю в исходном виде.
'''
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

Задание №3

```python
# Рассчитать количество пропусков для всех выбранных столбцов
df.isnull().sum()
```

```python
# Принять и обосновать решение о методе работы с пропусками по каждому столбцу
'''
Пропуски во всех столбцах - из-за недостатка знаний. 
Заполнение пропусков может привести к искажениию результатов. В рамках тренировки, можно заполнить пропуски средними значениями.
Пропуски в Surgery, Outcome, Mucous membranes, Temperature of extremities, Pain заполняем модальными значениями.
Rectal temperature заполняем медианой столбца.

'''
```

```python
df['Surgery'].fillna(df['Surgery'].mode()[0], inplace=True)
df['Outcome'].fillna(df['Outcome'].mode()[0], inplace=True)
df['Mucous membranes'].fillna(df['Mucous membranes'].mode()[0], inplace=True)
df['Temperature of extremities'].fillna(df['Temperature of extremities'].mode()[0], inplace=True)
df['Pain'].fillna(df['Pain'].mode()[0], inplace=True)
df['Rectal temperature'].fillna(df['Rectal temperature'].median(), inplace=True)
```
